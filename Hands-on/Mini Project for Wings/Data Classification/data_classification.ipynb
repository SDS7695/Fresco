```python
combi = pd.concat([train,test],axis=0).reset_index()
combi.head()
```

```python
def clean_df(df,col):
    final_df = pd.DataFrame()
    for i in col:
        dum_df = pd.get_dummies(df[i],drop_first = True)
        dum_df.columns = [i+'_'+str(x) for x in dum_df.columns]
        
        final_df = pd.concat([final_df,dum_df],axis=1)
    
    final_df = pd.concat([df,final_df],axis=1)
    final_df.drop(col,axis=1,inplace=True)
    return final_df
```

```python
col =[]
for i in train.columns:
    if train[i].dtype=='object':
        col.append(i)
```

```python
col
```

```python
combi_data = clean_df(combi,col)
```

```python
combi_data.head()
```

```python
train_data = combi_data[:43957]
test_data  = combi_data[43957:]
train_data.shape,test_data.shape
```

```python
test_data.head()
```

```python
X = train_data.drop(['income_>50K'],axis=1)
y = train['income_>50K']
test_data = test_data.drop(['income_>50K'],axis=1)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split,StratifiedKFold
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=1029)
```

```python
from sklearn import naive_bayes
model = naive_bayes.MultinomialNB()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict_proba(X_test) 
y_pred = y_pred[:,1] >= 0.3        # if prediction is greater than or equal to 0.3 than 1 else 0
y_pred = y_pred.astype(np.int)

print(accuracy_score(y_test,y_pred))
print('\n')
print(confusion_matrix(y_test,y_pred))
print('\n')
print(classification_report(y_test,y_pred))
```

```python
test_pred = model.predict_proba(test_data)
test_pred = test_pred[:,1] >= 0.3
test_pred = test_pred.astype(np.int)

pred = pd.DataFrame(test_pred,columns=['outcome'])
pred.to_csv('Wings_Classification.csv',index_label='id')
```
